```python
#Importamos las librerías pandas. numpy y matplotlib respectivamente.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from funpymodeling.exploratory import freq_tbl
```

```python
#Cargar achivo csv desde equipo
from google.colab import files
files.upload()
```

```python
df = pd.read_csv("Microretailer_limpio_evidencia2.csv") 
```

```python
freq_tbl(df)
```

```python
table1= freq_tbl(df['184_store_devices'])
table1
```

```python
#Obtengo un filtro de los valores más reelevantes de la variables categórica seleccionada
Filtro= table1[table1['frequency']>1]
Filtro
```

```python
#Ajusto el indice de mi dataframe
Filtro_index= Filtro.set_index('184_store_devices')
Filtro_index
```

```python
#Realizamos grafico de barras del dataframe filtrado
Filtro_index.plot(kind = 'bar', width=1, figsize=(10,4))
plt.title('Dispositivo de venta')
plt.xlabel('¿Qué dispositivo usa para el registro de sus ventas? ')
plt.ylabel('Frecuencia')
```

```python
table2= freq_tbl(df['189_payment_methods'])
table2
```

```python
#Obtengo un filtro de los valores más reelevantes de la variables categórica seleccionada
Filtro2= table2[table2['frequency']>1]
Filtro2
```

```python
#Ajusto el indice de mi dataframe
Filtro_index2= Filtro2.set_index('189_payment_methods')
Filtro_index2
```

```python
#Realizamos grafico de barras del dataframe filtrado
Filtro_index2.plot(kind = 'bar', width=1, figsize=(10,4))
plt.title('Método de pago preferido')
plt.xlabel('¿Con qué método de pago suelen hacer las compras más seguido?')
plt.ylabel('Frecuencia')
```
