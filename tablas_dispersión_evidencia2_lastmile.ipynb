```python
#Importamos las librerías pandas. numpy y matplotlib respectivamente.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from funpymodeling.exploratory import freq_tbl
```

```python
#Cargar achivo csv desde equipo
from google.colab import files
files.upload()
```

```python
df = pd.read_csv("lastmile_limpio_evidencia2 (1).csv") 
```

```python
freq_tbl(df)
```

```python
table1= freq_tbl(df['type_of_vehicle'])
table1
```

```python
#Obtengo un filtro de los valores más reelevantes de la variables categórica seleccionada
Filtro= table1[table1['frequency']>1]
Filtro
```

```python
#Ajusto el indice de mi dataframe
Filtro_index= Filtro.set_index('type_of_vehicle')
Filtro_index
```

```python
#Realizamos grafico de barras del dataframe filtrado
Filtro_index.plot(kind = 'bar', width=1, figsize=(10,4))
plt.title('Medio de transporte de lastmile')
plt.xlabel('¿En qué se hacen las entregas?')
plt.ylabel('Frecuencia')
```

```python
table2= freq_tbl(df['vehicles_unloading_door'])
table2
```

```python
#Obtengo un filtro de los valores más reelevantes de la variables categórica seleccionada
Filtro2= table2[table2['frequency']>1]
Filtro2
```

```python
#Ajusto el indice de mi dataframe
Filtro_index2= Filtro2.set_index('vehicles_unloading_door')
Filtro_index2
```

```python
#Realizamos grafico de barras del dataframe filtrado
Filtro_index2.plot(kind = 'bar', width=1, figsize=(10,4))
plt.title('Zona de descarga de los repartidores en los negocios')
plt.xlabel('¿Las descargas se hacen a un costado o al frente de la entrada del negocio?')
plt.ylabel('Frecuencia')
```
