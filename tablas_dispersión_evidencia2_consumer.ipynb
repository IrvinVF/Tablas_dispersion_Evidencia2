```python
#Importamos las librerías pandas. numpy y matplotlib respectivamente.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from funpymodeling.exploratory import freq_tbl
```

```python
#Cargar achivo csv desde equipo
from google.colab import files
files.upload()
```

```python
df = pd.read_csv("consumer_limpio_4 (2).csv") 
```

```python
freq_tbl(df)
```

```python
table1= freq_tbl(df['287_wire_transfers'])
table1
```

```python
#Obtengo un filtro de los valores más reelevantes de la variables categórica seleccionada
Filtro= table1[table1['frequency']>1]
Filtro
```

```python
#Ajusto el indice de mi dataframe
Filtro_index= Filtro.set_index('287_wire_transfers')
Filtro_index
```

```python
#Realizamos grafico de barras del dataframe filtrado
Filtro_index.plot(kind = 'bar', width=1, figsize=(10,4))
plt.title('Recargas telefónicas de los clientes')
plt.xlabel('¿En dónde hace recargas telefónicas?')
plt.ylabel('Frecuencia')
```

```python
table2= freq_tbl(df['309_wiretransfers_in_store'])
table2
```

```python
#Obtengo un filtro de los valores más reelevantes de la variables categórica seleccionada
Filtro2= table2[table2['frequency']>1]
Filtro2
```

```python
#Ajusto el indice de mi dataframe
Filtro_index2= Filtro2.set_index('309_wiretransfers_in_store')
Filtro_index2
```

```python
#Realizamos grafico de barras del dataframe filtrado
Filtro_index2.plot(kind = 'bar', width=1, figsize=(10,4))
plt.title('Interés de los clientes en hacer recargas')
plt.xlabel('¿Le gustaría hacer recargas en este establecimiento?')
plt.ylabel('Frecuencia')
```
